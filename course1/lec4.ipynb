```python
def new_abs_function(x):
    if x < 0:
        abs_value = -x
    else:
        abs_value = x
    return(abs_value)
```

```python
print(new_abs_function(5),new_abs_function(-5))
```

```python
def generate_data(n):
    eps_values = []
    for i in range(n):
        e = np.random.randn()
        eps_values.append(e)
    return eps_values

data = generate_data(100)
plt.plot(data)
plt.show()
```

```python
#Exercise 4.1

def factorial(n, f = lambda x: x):
    k = 1
    for i in range(f(n)):
        k = k*(i+1)
    return k

f = lambda x: x**2 + 1 if x%2 == 0 else x**2
factorial(2,f)
```
