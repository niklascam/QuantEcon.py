Version with a for loop

```python
ts_length = 100
eps_values = []

for i in range(ts_length):
    e = np.random.rand()
    eps_values.append(e)

plt.plot(eps_values)
plt.show()
```

Version with a while loop

```python
ts_length=100
eps_values=[]
i=0
while i < ts_length:
    e=np.random.randn()
    eps_values.append(e)
    i=i+1
plt.plot(eps_values)
plt.show()
```

```python
#Exercise 3.1
T=200
alfa=0.9

x = np.empty(T+1)
x[0] = 0
x_new=0

for t in range(T):
    x[t+1] = alfa*x[t] + np.random.randn()

plt.plot(x)
plt.show()
    
```

```python
#Exercise 3.2

T=200
alfa_st= [0, 0.8, 0.98]

for alfa in alfa_st:
    x = np.empty(T+1)
    x[0] = 0

    for t in range(T):
        x[t+1] = alfa*x[t] + np.random.randn()
    
    plt.plot(x, label=f'$\\alpha = {alfa}$')

plt.legend()
plt.show()
```

```python
#Exercise 3.3

T = 200
alfa=0.9
x = np.empty(T+1)
x[0]=0

for t in range(T):
    x[t+1] = alfa*np.abs(x[t]) + np.random.randn()

plt.plot(x)
plt.show()
```

```python
#Exercise 3.4

T = 200
alfa=0.9
x = np.empty(T+1)
x[0]=0

for t in range(T):
    if x[t] < 0:
        x[t+1] = alfa*(-1)*x[t] + np.random.randn()
    else:
        x[t+1] = alfa*x[t] + np.random.randn()

plt.plot(x)
plt.show()
```

```python
#Exercise 3.5

n = 10000
count = 0
for i in range(n):
    u,v = np.random.uniform(), np.random.uniform()

    d = np.sqrt((u - 0.5)**2 + (v - 0.5)**2)

    if d < 0.5:
        count += 1

area_estimate = count/n*4
print(area_estimate)
```
